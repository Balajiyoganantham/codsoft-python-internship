```python
import tkinter as tk
import random
import pygame
from pygame.locals import *

# Initialize Pygame
pygame.init()

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

# Choices
CHOICES = ["rock", "paper", "scissors"]

def generate_computer_choice():
    return random.choice(CHOICES)

def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!"
    elif (user_choice == "rock" and computer_choice == "scissors") or \
         (user_choice == "scissors" and computer_choice == "paper") or \
         (user_choice == "paper" and computer_choice == "rock"):
        return "You win!"
    else:
        return "Computer wins!"

def play_game(user_choice):
    computer_choice = generate_computer_choice()
    winner = determine_winner(user_choice, computer_choice)
    result_label.config(text=f"Computer chose: {computer_choice}. {winner}")

# Set up Tkinter window
root = tk.Tk()
root.title("Rock, Paper, Scissors")

# Set up Pygame screen
WIDTH, HEIGHT = 400, 300
screen = pygame.Surface((WIDTH, HEIGHT))

# Initialize fonts
font = pygame.font.Font(None, 24)

# Create Tkinter widgets
label = tk.Label(root, text="Choose: Rock, Paper, or Scissors")
label.pack()

result_label = tk.Label(root, text="")
result_label.pack()

rock_button = tk.Button(root, text="Rock", command=lambda: play_game("rock"))
rock_button.pack()

paper_button = tk.Button(root, text="Paper", command=lambda: play_game("paper"))
paper_button.pack()

scissors_button = tk.Button(root, text="Scissors", command=lambda: play_game("scissors"))
scissors_button.pack()

# Function to update Pygame display
def update_display():
    pygame_display = pygame.transform.scale(screen, (root.winfo_width(), root.winfo_height()))
    pygame_surface = pygame_display.convert()
    pygame_surface.set_colorkey((255, 255, 255))
    pygame_surface.fill((255, 255, 255))
    root_display = pygame_surface.get_rect()
    root_display.topleft = (0, 0)
    root.blit(pygame_surface, root_display)
    root.after(1000 // 60, update_display)

# Run Pygame event loop
def pygame_loop():
    for event in pygame.event.get():
        if event.type == QUIT:
            root.destroy()
            pygame.quit()
            return
    update_display()
    root.after(1000 // 60, pygame_loop)

# Run Tkinter main loop
root.after(1000 // 60, pygame_loop)
root.mainloop()
```
